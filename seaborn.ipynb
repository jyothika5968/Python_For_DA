```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df=pd.read_csv('/content/Telco-Customer-Churn.csv')
```

```python
df.head(5)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Churn', palette='Set2')
plt.title('Total Number of Retained vs. Churned Customers')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette='pastel')
plt.title('Distribution of Monthly Charges by Churn Status')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='tenure', hue='Churn', kde=True, multiple='stack', palette='coolwarm')
plt.title('Customer Tenure Distribution (Months)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Contract', hue='Churn', palette='viridis')
plt.title('Churn Rate Across Different Contract Types')
plt.xlabel('Contract Type')
plt.show()
```
