```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data ={'year':[1920,1930,1940,1950,1960,1970,1980,1990,2020],'Exchange rate':[65,67,68,69,70,71,72,73,75]}
```

```python
type(Data)
```

```python
len(Data)
```

```python
# convert dict into Data Frame
df=pd.DataFrame(Data)
```

```python
df
```

```python
df.plot(x='year',y='Exchange rate',kind='bar')
```

```python
df.plot(x='year',y='Exchange rate',kind='barh')
```

```python
# another method using matplotlib
plt.scatter(df['year'],df['Exchange rate'])
plt.xlabel('year')
plt.ylabel('Exchange rate')
plt.title('year vs Exchange rate')
```

```python
plt.pie(df['Exchange rate'],labels=df['year'])
plt.show()
```

```python
# Pie Chart
data ={'Tasks':[331,332,333]}
plt.pie(data['Tasks'],labels=['ongoing','completed','pending'])
plt.show()
```

```python
churn_df=pd.read_csv('/content/Telco-Customer-Churn.csv')
churn_df.head(5)
```

```python
churn_df.gender.value_counts().plot()
```

```python
churn_df['Churn'].value_counts().plot(kind='pie')
```

```python
plt.bar(churn_df['gender'],churn_df['TotalCharges'].max(),color='cyan')
plt.xlabel('gender')
plt.ylabel('TotalCharges')
plt.title('gender vs TotalCharges')
plt.show()
```

```python
plt.bar(churn_df['gender'],churn_df['tenure'].max(),color='blue', edgecolor='black')
plt.xlabel('gender')
plt.ylabel('tenure')
plt.title('gender vs tenure')
plt.show()
```

```python
# SCATTER PLOT
plt.scatter(churn_df['Partner'],churn_df['tenure'])
plt.xlabel('Partner')
plt.ylabel('tenure')
plt.title('gender vs Partner')
plt.show()
```

```python
# SCATTER PLOT
plt.scatter(churn_df['MonthlyCharges'],churn_df['tenure'])
plt.xlabel('MonthlyCharges')
plt.ylabel('tenure')
plt.title('gender vs MonthlyCharges')
plt.show()
```

```python
plt.hist(churn_df['tenure'])
plt.xlabel('tenure')
plt.ylabel('Frequency')
plt.title('tenure distribution')
plt.show()
```

```python
# Pie Chart
labels_ex ='Pavascript','Python','Python','R'
sizes_ex=[215,130,245,210]
colors_ex=['gold','yellowgreen','lightcoral','lightskyblue']
plt.pie(sizes_ex,labels=labels_ex,colors=colors_ex,autopct='%1.1f%%',shadow=True,startangle=140)
plt.axis('equal')
plt.show()
```

```python
labels_ex ='Pavascript','Python','Python','R'
sizes_ex=[215,130,245,210]
explode_labels=plt.subplot()
explode_labels.pie(sizes_ex,labels=labels_ex,autopct='%1.1f%%',shadow=True,startangle=140)
plt.show()
#
```

```python
labels=['javascript','java','python','r']
sizes=[15,20,10,25]
explode=[0.1,0.1,0.1,0.1]
fig1,ax1=plt.subplots()
ax1.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
ax1.axis('equal')
plt.show()
```
