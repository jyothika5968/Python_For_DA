```python
#given a positive no ngreater than 2 create a numpy array of size (n*n) with all zeros and ones such tht it forms a shape of PLUS

import numpy as np
def create_shape_of_array(n):
  arrayy=np.zeros((n,n),dtype=int)
  mid = n // 2
  arrayy[:,mid]=1
  arrayy[mid,:]=1

  return arrayy
n=7
result=create_shape_of_array(n)
print(result)
```
